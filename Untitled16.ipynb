```python
!pip install pandas seaborn matplotlib scikit-learn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
from google.colab import files
uploaded = files.upload()
df = pd.read_excel(next(iter(uploaded.keys())))  # Read the first uploaded file
df = pd.read_excel("OnlineRetail (1) (1).xlsx")
print(df.head())
print(df.info())
print(df.isnull().sum())
df.dropna(subset=['Description', 'CustomerID'], inplace=True)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

top_products = df.groupby('Description')['Quantity'].sum().nlargest(10)
sns.barplot(x=top_products.index, y=top_products.values)
plt.xticks(rotation=90)
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Products')
plt.ylabel('Quantity Sold')
plt.show()
# Step 6: Creating a recommendation system based on the most purchased items
# Create a pivot table for products purchased by customers
purchase_matrix = df.pivot_table(index='CustomerID', columns='Description', values='Quantity', fill_value=0)

# Calculate cosine similarity between customers
cosine_sim = cosine_similarity(purchase_matrix)

# Create a DataFrame from the similarity matrix
similarity_df = pd.DataFrame(cosine_sim, index=purchase_matrix.index, columns=purchase_matrix.index)

# Global Popular
global_popular_items = df.groupby('Description')['Quantity'].sum().nlargest(10)

# Viz
plt.figure(figsize=(10,5))
sns.barplot(x=global_popular_items.index, y=global_popular_items.values)
plt.xticks(rotation=90)
plt.title('Top 10 Popular Items Globally')
plt.xlabel('Items')
plt.ylabel('Quantity Sold')
plt.show()

# Country-wise Popular
country_popular_items = df.groupby(['Country', 'Description'])['Quantity'].sum().reset_index()

# Top item for each country
country_top_items = country_popular_items.loc[country_popular_items.groupby('Country')['Quantity'].idxmax()]

# Viz
plt.figure(figsize=(18, 9))
sns.barplot(x='Quantity', y='Country', data=country_top_items, hue='Description', dodge=False)
plt.title('Top Item Sold Countrywise')
plt.xlabel('Quantity Sold')
plt.ylabel('Country')
plt.legend(title='Items', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Month from InvoiceDate
df['Month'] = df['InvoiceDate'].dt.to_period('M')

# Popular Items Month-wise
monthly_popular_items = df.groupby(['Month', 'Description'])['Quantity'].sum().reset_index()

# Top item for each month
monthly_top_items = monthly_popular_items.loc[monthly_popular_items.groupby('Month')['Quantity'].idxmax()]

# Viz
plt.figure(figsize=(10, 5))
sns.barplot(x='Month', y='Quantity', data=monthly_top_items, hue='Description', dodge=False)
plt.title('Top Item Sold Month-wise')
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(title='Items', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

def analyze_recommendations (customer_id, n_recommendations=5):
    # Check if customer ID is in the DataFrame
    if customer_id not in purchase_matrix.index:
        return "Customer ID not found."

    # Get similar customers
    similar_customers = similarity_df[customer_id].sort_values(ascending=False)
    similar_customers = similar_customers[similar_customers.index != customer_id]  # Exclude self

    # Get top N similar customers
    top_customers = similar_customers.head(n_recommendations).index

    # Get the products purchased by these customers
    recommended_products = df[df['CustomerID'].isin(top_customers)]['Description'].value_counts().head(n_recommendations)

    # Predict which items the customer might like based on purchasing patterns
    predictions = recommended_products.index.tolist()

    # Print results
    print(f"Recommendations for Customer ID {customer_id}:")
    for idx, product in enumerate(predictions, start=1):
        print(f"{idx}. {product}")

    return predictions

    # Example usage
recommendation = analyze_recommendations (17850)
```
